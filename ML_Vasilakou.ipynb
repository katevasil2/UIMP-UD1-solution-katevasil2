```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

# Normal distribution
---




```python

n = 1000

data_normal = np.random.normal(loc=0, scale=1, size=n)

mean = np.mean(data_normal)
std = np.std(data_normal)
print(mean, std)
theor_ci = (mean - 1.96*std/np.sqrt(n), mean + 1.96*std/np.sqrt(n))
print("95% CI:", theor_ci)
```

```python
plt.hist(data_normal, bins=30, alpha=0.7, color='skyblue', edgecolor='black')

#Mean line
plt.axvline(mean, color='red', linestyle='-', linewidth=2, label=f'Mean = {mean:.2f}')

#CI limits
plt.axvline(theor_ci[0], color='green', linestyle='--', linewidth=2, label=f'CI lower = {theor_ci[0]:.2f}')
plt.axvline(theor_ci[1], color='green', linestyle='--', linewidth=2, label=f'CI upper = {theor_ci[1]:.2f}')

plt.title("Histogram of Normal Distribution Data with Mean and 95% CI")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
def hoeffding_bound(n, epsilon, a, b):
    return 2 * np.exp(-2 * n * epsilon**2 / (b - a)**2)

#Randomly selected epsilon
epsilon = 0.001

a = np.min(data_normal)
b = np.max(data_normal)

# Calculate Hoeffding bound
bound = hoeffding_bound(n, epsilon, a, b)
print("Min value (a):", a)
print("Max value (b):", b)
print("Hoeffding bound:", bound)
```

```python
def chebyshev_bound(data, k):
    return np.mean(np.abs(data - mean) >= k * std), 1/k**2

# Example for normal data
k = 2
empirical_prob, cheb_theoretical = chebyshev_bound(data_normal, k)
print(f"Empirical probability: {empirical_prob}, Chebyshev bound: {cheb_theoretical}")
```

---

# Beta distribution
---

```python
n = 1000

# Beta distribution
alpha = 2
beta = 5

data_beta = np.random.beta(a=alpha, b=beta, size=n)

# Compute sample mean and std
mean = alpha / (alpha + beta)
std = np.sqrt(alpha * beta / ((alpha + beta)**2 * (alpha + beta + 1)))

# 95% CI for the sample mean
theor_ci = (mean - 1.96*std/np.sqrt(n), mean + 1.96*std/np.sqrt(n))
print("Mean:", mean)
print("Std:", std)

# 95% confidence interval
theor_ci = (mean - 1.96*std/np.sqrt(n), mean + 1.96*std/np.sqrt(n))
print("95% CI:", theor_ci)
```

```python
plt.hist(data_beta, bins=30, alpha=0.7, color='skyblue', edgecolor='black')

#Mean line
plt.axvline(mean, color='red', linestyle='-', linewidth=2, label=f'Mean = {mean:.2f}')

#CI limits
plt.axvline(theor_ci[0], color='green', linestyle='--', linewidth=2, label=f'CI lower = {theor_ci[0]:.2f}')
plt.axvline(theor_ci[1], color='green', linestyle='--', linewidth=2, label=f'CI upper = {theor_ci[1]:.2f}')

plt.title("Histogram of Beta Distribution Data with Mean and 95% CI")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

# Comments

Chebyshevs Inequality: This inequality works (gives a result that can e interpreted as probability) for all the
distributions since they all have finite variance (since we take a sample of 1000). I used k=2 which should lead to much smaller results (especially for the normal distribution), reflecting that this bound is correct but extremely loose.

Hoeffdings Inequality: This bound works well for bounded distributions. So this bound can be trusted for the unifrom and beta distributions but not for the normal. I did not keep it in [0,1] to emphasize that it does not work for the normal distribution.

 689599.7 rule: This rule derives from the Gaussian distribution and applies only to that. The reason why it does not apply to beta and uniform can be seen just by the shape of the distributions. Uniform has is a didtribution with large variance but bounded by 0 and 1. That leads to finding all the values (100%) within the range of 2 sigmas, so the rule already collapses. Beta distribution is closer to this rule but doen not work mainly because is not symmetric.
